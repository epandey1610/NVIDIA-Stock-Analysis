```python
import pandas as pd

import yfinance as yf
```

```python
nvda = yf.Ticker("NVDA")
```

```python
nvda = nvda.history( period = "max")
```

```python
nvda.plot.line(y="Close", use_index = True)
```

```python
del nvda["Dividends"]

del nvda["Stock Splits"]
```

```python
nvda.head()
```

```python
#checking data types and covert them to numerical form

print(nvda.dtypes)

# Convert relevant columns to numerical types
for col in ['Open', 'High', 'Low', 'Close', 'Volume']:
    nvda[col] = pd.to_numeric(nvda[col], errors='coerce')

nvda.dtypes
```

```python
nvda["Tomorrow"] = nvda["Close"].shift(-1)
```

```python
nvda.head()
```

```python
# Daily Price Change

nvda["Price_Change"] = nvda["High"] - nvda["Close"]
```

```python
nvda.head()
```

```python
# Now creating a Target

# Compare 'Tomorrow' with 'Close' and convert the boolean result to integers
nvda["Target"] = (nvda["Tomorrow"] > nvda["Close"]).astype(int)

# Verify the change
print(nvda["Target"].head())
```

```python
nvda
```

```python
# Our data column is set under Date and Time so we would change as we don't look for hourly changes in stock as it could be for later.

# Reset the index to make "Date" a column

nvda.reset_index(inplace = True)

# Modifying the "Date" column to only include the date part

nvda["Date"] = nvda["Date"].dt.date

#Optionally, Setting the date column back to as the index

nvda.set_index("Date", inplace=True)
```

```python
nvda
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, min_samples_split=100, random_state=1)

train = nvda.iloc[:-100]
test = nvda.iloc[-100:]

predictors = ["Close", "Volume", "Open", "High", "Low", "Price_Change"]

model.fit(train[predictors], train["Target"])
```

```python
from sklearn.metrics import precision_score

preds = model.predict(test[predictors])
```

```python
preds = pd.Series(preds, index=test.index)
```

```python
preds
```

```python
precision_score(test["Target"], preds)
```

```python
combined = pd.concat([test["Target"], preds], axis = 1)
```

```python
combined.plot()
```

```python
#creating Back testing loop

def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
# Every year has 250 trading days,therefore, 250*10=2500
# Which suggests a decade long data

def backtest(data, model, predictors, start=2500, step=250):
  all_predictions = []
  for i in range(start, data.shape[0],step):
    train = data.iloc[0:i].copy()
    test = data.iloc[i:(i+step)].copy()
    predictions = predict(train, test, predictors, model)
    all_predictions.append(predictions)

  return pd.concat(all_predictions)
```

```python
predictions = backtest(nvda, model, predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
# 0 = Suggesting market would go down  for 2222 days

# 1 = Suggesting market would go up for 1821 days
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
predictions["Target"].value_counts() / predictions.shape[0]
```

```python
# Accuracy is poor again
# Hence, we will use rolling Averages

horizions = [2,5,60,250,1000]
new_predictors = []

for horizion in horizions:
  rolling_averages = nvda.rolling(horizion).mean()

  ratio_column = f"Close_Ratio_{horizion}"
  nvda[ratio_column] = nvda["Close"] / rolling_averages["Close"]
  new_predictors.append(ratio_column)

  trend_column = f"Trend_{horizion}"
  nvda[trend_column] = nvda.shift(1).rolling(horizion).sum()["Target"]
  new_predictors.append(trend_column)

  new_predictors += [ratio_column, trend_column]
```

```python
nvda.head()
```

```python
# Removing Nan

nvda = nvda.dropna()

nvda
```

```python
# Updating the model

model = RandomForestClassifier(n_estimators=200, min_samples_split=50, random_state=1)
```

```python
# Adjusting the prediction filter

def predict(train, test, predictors, model):
  model.fit(train[predictors], train["Target"])
  preds = model.predict_proba(test[predictors])[:,1]
  preds[preds >= .6] = 1
  preds[preds < .6] = 0
  preds = pd.Series(preds, index=test.index, name="Predictions")
  combined = pd.concat([test["Target"], preds], axis=1)
  return combined
```

```python
predictions = backtest(nvda, model, new_predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```
